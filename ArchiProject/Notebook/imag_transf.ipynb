```python
def transform(img):

  import cv2
  
  img_tr = cv2.resize(img, (256, 256))
  # Convertimos a escala de grises
  img_canny = cv2.Canny(img_tr, 400, 50) 
  #invertir
  img_final = cv2.bitwise_not(img_canny)
  
  cv2.imwrite('prueba.jpg', img_final)

  tra_img = cv2.imread('prueba.jpg')

  conc = cv2.hconcat([img_tr, tra_img])
  return conc
```

```python
a = transform(imagen)
```

```python
import os

nombre_carpeta = '/content/drive/MyDrive/universidad_de_chile/imag_home/socal2/socal_pics'

contenido = os.listdir(nombre_carpeta)
cont = 1
for elemento in contenido:
    photo = os.path.join(nombre_carpeta, elemento)

    imagen = cv2.imread(str(photo))

    a = transform(imagen)


    cv2.imwrite(f"/content/drive/MyDrive/universidad_de_chile/img_dataset/{cont}.jpg",a)

    cont = cont + 1
```
